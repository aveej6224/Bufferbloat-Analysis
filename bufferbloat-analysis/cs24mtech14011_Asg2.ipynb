## Background

#### TCP Congestion Window Size
The TCP congestion window size parameter, typically styled "cwnd," is maintained by the sender and determines how much traffic can be outstanding (sent but not acknowledged) at any time. There are many algorithms for controlling the value of cwnd during a TCP connection, all with the goal of maximizing the connection's throughput while preventing congestion. The additive increase and multiplicative decrease algorithm was discussed in lecture.

#### Bufferbloat
Bufferbloat is a phenomenon that happens when a switching device is configured to use excessively large buffers, which can in turn cause high latency and packet delay variation (jitter). This can happen even in a typical home network like the following:

```python
from mininet.topo import Topo

class BBTopo(Topo):
    "Simple topology for bufferbloat experiment."

    def __init__(self, queue_size):
        super(BBTopo, self).__init__()
        
        # Create switch s0 (the router)
        self.addSwitch('s0')
        
        # TODO: Create two hosts with names 'h1' and 'h2'
        h1 = self.addHost('h1',ip = '10.0.0.1')
        h2 = self.addHost('h2',ip = '10.0.0.2')
        
        # TODO: Add links with appropriate bandwidth, delay, and queue size parameters. 
        #       Set the router queue size using the queue_size argument
        #       Set bandwidths/latencies using the bandwidths and minimum RTT given in the network diagram above

        self.addLink(h1, 's0', bw=10, delay='10ms', max_queue_size=queue_size)
        self.addLink(h2, 's0', bw=10, delay='10ms', max_queue_size=queue_size)
        
        return
```

Next, we need a couple of helper functions to generate traffic between the two hosts. The following function starts a long-lived TCP flow which sends data from h1 to h2 using **iperf**.
```python
def start_iperf(net, experiment_time):
    # Start a TCP server on host 'h2' using perf. 
    # The -s parameter specifies server mode
    # The -w 16m parameter ensures that the TCP flow is not receiver window limited (not necessary for client)
    print ("Starting iperf server")
    h2 = net.get('h2')
    server = h2.popen("iperf -s -w 16m", shell=True)
    
    # TODO: Start an TCP client on host 'h1' using iperf. 
    #       Ensure that the client runs for experiment_time seconds
    print("Starting iperf client")
    h1 = net.get('h1')
    client_command = "iperf -c 10.0.0.2 -t {}".format(experiment_time)
    client = h1.popen(client_command, shell=True)

    # Optionally, wait for the client to finish
    client.wait()
```

Next, we develop some helper functions to measure the congestion window of the TCP traffic. This will let us analyze at the dynamics of the TCP connections in the mininet network. The following functions are already complete.

```python
from subprocess import Popen
import os

def start_tcpprobe(outfile="cwnd.txt"):
    print("Starting TCP probe. Output: {}".format(outfile))
    os.system("sudo rmmod tcp_probe; sudo modprobe tcp_probe")
    process = Popen("sudo cat /proc/net/tcpprobe > " + outfile, shell=True)


def stop_tcpprobe():
    Popen("killall -9 cat", shell=True).wait()
```

Finally, we need a helper function that runs on h2, fetches the website from h1 every 3 seconds for `experiment_time`, and prints the average and standard deviation of the download times. This function is already complete

```python
from time import time
from numpy import mean, std
from time import sleep

def fetch_webserver(net, experiment_time):
    h2 = net.get('h2')
    h1 = net.get('h1')
    download_times = []
    
    start_time = time()
    while True:
        sleep(3)
        now = time()
        if now - start_time > experiment_time:
            break
        fetch = h2.popen("curl -o /dev/null -s -w %{time_total} ", h1.IP(), shell=True)
        download_time, _ = fetch.communicate()
        print ("Download time: {0}, {1:.1f}s left...".format(download_time, experiment_time - (now-start_time)))
        download_times.append(float(download_time))
        
    average_time = mean(download_times)
    std_time = std(download_times)
    print ("\nDownload Times: {} sec average, {} sec stddev\n".format(average_time, std_time))
```

#### A2.
The components that contribute to RTT: propagation delay, transmission delay, and queuing delay.  
Let:
- $q$ represent the number of packets in the queue.
- $L$ be the packet size in bits.
- $R$ be the link transmission rate in bits per second.
- $d_{prop}$ represent the one-way propagation delay.

The RTT consists of the time taken for a packet to travel from source to destination and back, including the delay due to packets in the queue. Hence, the RTT can be expressed as:

$$
RTT = 2 \cdot d_{prop} + \frac{q \cdot L}{R}.
$$

Here:
1. $2 \cdot d_{prop}$ accounts for the propagation delay for the round trip.
2. $\frac{q \cdot L}{R}$ represents the queuing delay, where each packet in the queue contributes to the overall delay.

This equation generalizes RTT as a function of the queue size $q$, packet size $L$, and link properties $R$ and $d_{prop}$.

#### A5.

The bufferbloat effect is a general phenomenon and not specific to a particular network type, traffic pattern, or TCP congestion control algorithm. It occurs whenever buffers are large enough to cause excessive queuing delays during congestion. While its severity may vary based on the algorithm (ex: CUBIC vs. Reno), the fundamental issue is universal.

**When larger buffers help:**  
Increasing buffer size can improve performance in networks with high bandwidth-delay products, such as satellite links, where the delay is intrinsic and stable. Larger buffers help prevent underutilization by accommodating bursts of data and ensuring the link remains saturated.

**Why its rare:**  
In most typical networks, excessive buffering causes more harm than good by introducing high delays and degrading responsiveness, especially for real-time or interactive applications.


#### Q6.
Identify and describe a way to mitigate the bufferbloat problem without reducing buffer sizes.  

#### A6.

One effective way to mitigate bufferbloat without reducing buffer sizes is to use **Active Queue Management (AQM)** algorithms, such as **CoDel (Controlled Delay)** or **RED (Random Early Detection)**. These algorithms manage queue behavior proactively, ensuring that packets are dropped or marked before buffers become excessively full.

**How AQM works:**  
1. AQM monitors queue lengths or delays.
2. When congestion is detected, it either drops packets or marks them with Explicit Congestion Notification (ECN) to signal the sender to reduce the transmission rate.

**Why this helps:**  
By controlling queue growth, AQM ensures lower queuing delays, preventing excessive RTT increases and maintaining responsiveness. This approach avoids the inefficiencies of large queues while preserving sufficient buffer space to handle traffic bursts.
